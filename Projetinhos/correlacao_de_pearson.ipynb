### CORRELAÇÃO DE PEARSON

o coeficiente de correlação de Pearson é uma tecnica algébrica que mede o grau da correlação entre DUAS VARIÁVEIS de escala métrica, esse coeficiente é representado por ρ (em grego ró) e o mesmo assume apenas valores entre -1 e 1

#### REPRESENTAÇÃO DA CORRELAÇÃO DE PEARSON EM PYTHON

```python
# importando as bibliotecas necessárias

import numpy as np
import matplotlib.pyplot as plt
```

```python
# usando dois tipos de series 

x = [2,9,5,6,7,8]
y = [5,3,7,5,9,7]

# após as váriaveis serem declaradas, é usado o 
# método CORRCOEF da biblioteca NUMPY

correlaçãoTotal = np.corrcoef(x, y)[0, 1]
```

```python
# criando o gráfico de dispersão usando a biblioteca
# mathplotlob

plt.scatter(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('dispersão')

# adicionando o coeficiente como texto no gráfico
plt.text(1, 5, f'Correlação de pearson: {correlaçãoTotal:.2f}', fontsize=12)

plt.show()
```
