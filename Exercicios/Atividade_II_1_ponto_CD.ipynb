### Dados obtidos do cliente. A lista abaixo possui uma tupla com a idade e a somatória dos gastos em um determinado serviço  

```python
# SÓ EXECUTAR A CÉLULA, NÃO PRECISA ALTERAR
# Importa biblioteca para usar números aleatórios
import random as numero_aleatorio # a palavra 'as' indica que queremos dar um apelido a essa biblioteca
# Defini o tamanho da população
quantidade_pessoas = numero_aleatorio.randint(1000, 5000)
```

Aqui será feito o preenchimento da lista com idade e gasto por idade (SÓ EXECUTAR)

```python
# SÓ EXECUTAR A CELULA, NÃO PRECISA ALTERAR
# Preenchimento da lista dos dados com números aleatórios. 
lista = []
for indice in range(quantidade_pessoas):
  idade = numero_aleatorio.randint(18, 40) # Gera valores entre 18 e 40
  salario = numero_aleatorio.randint(2800, 10000) # Gera valores entre 2800 até 10000
  lista.append((idade, salario))
print(lista)
```
