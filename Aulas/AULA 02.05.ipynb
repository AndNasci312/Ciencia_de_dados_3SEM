Numpy

é uma biblioteca que fornece um objeto de matriz bidimencional

o numpy oferece uma enorme variedade de maneiras rápidas e eficientes de criar martizes e manipular dados numéricos dentro elas;

as matrizes numpy são mais rápidas e compactas do que as listas python...



```python
import numpy as np
```

```python
# chamando a função array dentro de uma matriz unidimencional 
a = np.array([1, 2, 3])
print(a)
```

```python
# chamando a função array dentro de uma matriz bidimencional
a = np.array([[1, 2, 3],[2, 5, 7]])
print(a)
```

```python
vetor = np.zeros(2)
matriz = np.zeros((2,3))
print(vetor) 
print('------------')
matriz
```

```python
vetor = np.ones(2, dtype=np.int64)
matriz = np.ones((5, 5), dtype=np.float32)
print(vetor)
print(matriz)
```

```python
# para usar o empty, basta passar o vetor ou matriz entre os parênteses, assim ele irá criar uma matriz proxima de zero
vetor = np.empty(2)
matriz = np.empty((2,5))
print(vetor)
print(matriz)
```

```python
# para usar o arrange, basta o valor de até quando criar os itens 
vetor = np.arange(4)
matriz = vetor.reshape(2,2)
print(vetor)
print(matriz)
```

```python
vetor = np.linspace(0, 9, num=4)
print(vetor)
matriz = vetor.reshape(2,2)
print(matriz)
```
