```python
# Requisição dos dados e a atribuição que resulta a variável RESPOSTA
# link do aqrquivo CSV de exemplo: https://raw.githubusercontent.com/luiscarlosjunior/aulas-graduacao/master/data-science/analise-dados/datasets/csv/resumoAlunos.csv

import requests
url = "https://raw.githubusercontent.com/luiscarlosjunior/aulas-graduacao/master/data-science/analise-dados/datasets/csv/resumoAlunos.csv"
resposta = requests.get(url)
print(resposta)
```

```python
# SEPARANDO OS DADOS DO ARQUIVO CSV POR LINHAS

# armazenando os dados em uma variável de texto
texto = resposta.text
# separando por conjuntos únicos 
linhas = texto.split('\n')
# printando a quantidade de linhas e seu tipo
print(linhas)
print(type(linhas))
print(linhas[0])
print(linhas[-1])
```

```python
#retirar a ultima linha EM BRANCO do cojunto
print(linhas[-1])
del linhas[-1]
print(linhas[-1])
```

```python
# transformando linha em lista, primeiramente separando a lista atribuida por ;
linhas_lista = linhas[0].split(";")
print(linhas_lista)
print(type(linhas_lista))

# mostrando os dados únicos
print(linhas_lista[6])
print(linhas_lista[5])

# retirando marcações indesejadas
print(linhas_lista[0])
linhas_lista[0] = linhas_lista[0].lstrip('\ufeff')
```
