Ajuste de Curvas

é uma técnica amplamente usada em diversas áreas da ciência, incluindo física, engenharia, biologia, economia e muitas outras;

é uma ferramenta importante porque permite encontrar uma relação matemática que descreve os dados de uma maneira simples e compacta;

o ajuste de curvas pode ser usado em previsões, finanças, modelagem, análise de tendencias, interpolação

e na maioria das vezes, o ajuste de curvas sempre é feito em relação a duas variáveis


variáveis fortes: com dados precisos 
variáveis fracas: com dados imprecisos e incosistentes 

variaveis de confusão

pode afetar a relação entre duas varíaveis, uma terceira variável, o processo de correção é conhecido como correção, diminuição de ruidos ou pre processamento 



```python
import numpy as np

x = [2,5,6,9,7,8,9,10]
y = [5,6,9,3,1,5,6,9]

alpha1, alpha2 = np.polyfit(x, y, 1)

print(f'a equação da curva é {alpha1} {alpha2}')
```

Coeficiente de pearson


essa tecnica é só adequada para avaliar a correlação linear entre duas variáveis

```python
# exemplo de uma implementação da correlação de pearson com numpy
x = np.array(x)
y = np.array(x)

corr_coef = np.corrcoef(x, y) [0] [1]
print(corr_coef)
```
