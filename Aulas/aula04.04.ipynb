```python
import requests
```

```python
url = "https://www.google.com"
conteudo = requests.get(url)
```

```python
print("")
print(conteudo)
print("")
print(conteudo.content)
print("")
print(conteudo.headers)
print("")
print(conteudo.status_code)


status = conteudo.status_code
if conteudo.status_code >= 200 and conteudo.status_code <= 299:
  print('requisição feita!')
else:
  print(f'verifique o codigo {status}')
```

```python
url = 'https://www.pudim.com'
dados = {
    "name": "john dows",
    "email": "johnson@don.com"
}

# resposta recebe request.post + a url e os dados pesquisados
# STATUS reebe resposta e o código de status do site

resposta = requests.post(url, dados)
status = resposta.status_code

if status >= 200 and status <= 299:
  print("o usuario foi adcionado com sucesso")
elif status >= 400 and status <= 499:
  print("erro no lado do cliente")
elif status >= 500 and status <= 599:
  print("erro no lado do servidor")

```

```python
url = "https://raw.githubusercontent.com/luiscarlosjunior/aulas-graduacao/master/data-science/analise-dados/datasets/csv/resumoAlunos.csv"

resposta = requests.get(url)
#pega cada linha do dataset
lista_itens = resposta.text.split("\n")

print("")
# exibe a quantidade de itens 
print(len(lista_itens))
# exibe os itens da lista
print(lista_itens[2])


```

```python
dicionario_alunos = dict()

tamanho = len(lista_itens)

for indice in range(tamanho):
  itens = lista_itens[indice].split(";")
  dicionario_alunos[indice] = itens
```

```python

soma = 0
tamanho = len(lista_itens)
for media in range(tamanho - 2):
  soma = soma + float(dicionario_alunos.get(media + 1)[6])
print(f'a média é {soma / tamanho}')
```
